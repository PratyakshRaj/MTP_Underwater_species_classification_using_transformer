```python
mat=[[32,0,0,0,3],[0,110,0,0,0],[0,0,69,2,3],[0,0,4,60,1],[0,0,0,0,71]]
```

```python
import numpy as np

def f1_from_cm(cm):
    cm = np.array(cm)
    num_classes = cm.shape[0]

    TP = np.diag(cm)
    FP = np.sum(cm, axis=0) - TP
    FN = np.sum(cm, axis=1) - TP

    # Avoid divide-by-zero
    precision = np.where(TP + FP == 0, 0, TP / (TP + FP))
    recall = np.where(TP + FN == 0, 0, TP / (TP + FN))

    f1 = np.where(
        precision + recall == 0,
        0,
        2 * precision * recall / (precision + recall)
    )

    macro_f1 = np.mean(f1)
    weighted_f1 = np.average(f1, weights=np.sum(cm, axis=1))

    return precision, recall, f1, macro_f1, weighted_f1

# Example usage:
prec, rec, f1, macro_f1, weighted_f1 = f1_from_cm(mat)

print("Per-class F1:", f1)
print("Macro F1:", macro_f1)
print("Weighted F1:", weighted_f1)
```
